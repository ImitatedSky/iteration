```python
input_a=5
input_b=1

a_w=2
b_w=-1

lr=0.01
res = 13

# dp(0)=[a_w, b_w, input_a*a_w+input_b*b_w]
f = [0 for i in range(100)]
def dp_f(n):
  if n <=0:
    f[n] = [a_w, b_w, input_a*a_w+input_b*b_w]
    return f[n]
  f[n] = [f[n-1][0]-(lr*(-2*input_a*(res-f[n-1][2]))),
          f[n-1][1]-(lr*(-2*input_b*(res-f[n-1][2]))),
          input_a*(f[n-1][0]-(lr*(-2*input_a*(res-f[n-1][2])))) + input_b*(f[n-1][1]-(lr*(-2*input_b*(res-f[n-1][2])))),
          ]
  return f[n]

for i in range(20):
  print(i,dp_f(i))
```

```python
input_a=5
input_b=1
input_c=-1

a_w=2
b_w=-1
c_w=-1

lr=0.01
res = 13

# dp(0)=[a_w, b_w, input_a*a_w+input_b*b_w]
f = [0 for i in range(100)]
def dp_f(n):
  if n <=0:
    f[n] = [a_w, b_w,c_w, input_a*a_w+input_b*b_w+input_c*c_w]
    return f[n]
  f[n] = [f[n-1][0]-(lr*(-2*input_a*(res-f[n-1][3]))),
          f[n-1][1]-(lr*(-2*input_b*(res-f[n-1][3]))),
          f[n-1][2]-(lr*(-2*input_c*(res-f[n-1][3]))),
          input_a*(f[n-1][0]-(lr*(-2*input_a*(res-f[n-1][3])))) + input_b*(f[n-1][1]-(lr*(-2*input_b*(res-f[n-1][3])))) + input_c*(f[n-1][2]-(lr*(-2*input_c*(res-f[n-1][3])))),
          ]
  return f[n]

for i in range(20):
  print(i,dp_f(i))
```
